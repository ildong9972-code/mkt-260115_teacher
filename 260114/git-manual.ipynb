# git 사용방법
1) 로컬컴퓨터 (여러분 컴퓨터) : git 응용프로그램 설치
2) 터미널 : git -v 명령어
- 검색바 > cmd > 클릭 : 검정색
- cmd + space > terminal
3) 만약, 없다면 > 구글 크롬 > git 다운로드 후 설치 : https://git-scm.com/
- homebrew : 대리로 응용프로그램 다운로드 후 설치해주는 중간 프로그램 -> 맥용
- chocholatey : 대리로 응용프로그램 다운로드 후 설치해주는 중간 프로그램 -> 윈도우용
4) 깃의 소유주 확인 & 변경
- git config --list
5) github 회원가입 : 계정정보 값으로 입력 : https://github.com/
- git : 나의 로컬컴퓨터에서 버전 관리 툴
- github : 나의 로컬컴퓨터에서 관리하고 있는 버전을 타인과 공유해서 함께 사용하고자 할 때, 혹은 내가 사용하고 있는 디바이스가 2대 이상인 경우, A에서도 데이터관리, B에서도 데이터관리
- repository : github > project에 따른 개별 폴더
- user.email : sejinpark@tcocs.com
- user.name : Divjason
- 터미널 재접속 > 관리자모드 : 전역(global)세팅
- git config --global user.name "Divjason"
- git config --global user.email "sejinpark@tcocs.com"
6) git commit & push
- cd "현재나의 루트폴더" 이동
- git 초기화 : git init
