# 파이썬 기초문법
- 출력 : print() -> 내장함수 = built in function / 내장 : 애초에 파이썬 문법 탄생될 때부터 존재
- 입력 : input("당신의 이름은?") -> 사용자로부터 어떤 값을 입력받을 수 있음 / placeholder = 입력해야할 값의 정보
    - 해당함수를 통해서 수집한 값은 무조건 "문자열"이 된다는 사실 기억!
- 연산 :
    - 사칙연산 : + @ -> + - * / % ** -> '+' : 문자간의 연결연산자로서도 사용가능하다!
    - 비교연산 : <, >, <=, >=, ==, != ("좌항 = 우항, 우항의 값을 좌항으로 삽입, 적용) 
- 변수 :
    - variable => 변할 수 있는 값을 넣는 공간 = 그릇
    - 변수명 작명 => 직관.명시 -> 누가봐도 한눈에 알아볼 수 있는 변수명 / 숫자가 먼저올 수 없음 (중간에 오는건 가능)
    - 팁 => 스네이크 표기법 (뱀 -> case_study / 카멜 표기법 -> caseStudy)
- 자료형 : 변수안에 담을 수 있는 값의 형태 -> 문자열 (String) / 정수 (Int) / 실수 (float) / 논리형 (Boolean)
    - 형변화 : 자동형변환 : input() / 명시적형변환 : 수동적으로 형태를 변환
    - 자료형확인 : type()
    - int() : 정수형 변환
    - float() : 실수형 변환
    - str() : 문자열 변환
    - bool() : 논리형 변환
    - list() : 리스트 자료구조 변환
    - dict() : 딕셔너리 자료구조 변환
    - tuple() : 튜플 자료구조 변환
    - set() : 세트(집합) 자료구조 변환
- 문자열 :
    - count() : 특정 문자열에서 어떤 단어가 몇 번 사용되었는지를 집계하고자 할 때 (대.소문자 구분)

```python
print('Hello World')
print(1+2)
print((1+2)*3)
```

```python
num1 = 4 # 변수 : 어떤 값을 담을 수 있는 그릇 -> 값을 언제든지 바꿔서 담을 수도 있다
num2 = 2 # 변수안에 값을 최초로 담았을 때, "할당"

# num1 = 8 # 변수안에 값을 "재할당"
print(num1 + num2) # 4, 2 -> 피연산자, + -> 연산자
print(num1 - num2)
print(num1 * num2)
print(num1 / num2) # 나누기 연산에 대한 몫
print(num1 % num2) # 나누기 연산에 대한 나머지 -> 홀수, 짝수 찾아야내야하는 상황
print(num1 ** num2)
```

```python
str01 = "David"
str02 = 'Jane'
int01 = 1
int02 = "1"
flo01 = 3.14
bool01 = True
bool02 = False

print(str01, str02, int01, int02)
print(type(int02)) # type 함수 : 인자값으로 입력된 변수안에 담긴 자료의 타입.형태를 식별.확인하고자 할 때 함수
print(type(flo01)) # alt + shift + 아래방향 화살표 = duplicate = 복제됨!!!
print(type(int01))
print(type(bool01))
```

```python
# 주피터 노트북 초심자 좋은이유 : 실제 코드 작성 외 굳이 외적인 불필요한 작업 x
# 주피터 x / scrapy 기법 : cursor AI
print(num1 > num2)
print(num1 < num2)
print(num1 >= num2)
print(num1 <= num2)
print(num1 == num2)
print(num1 != num2)
```

```python
radius = 10 # 반지름
pi = 3.1415 # 파이 = 원주율

radius2 = radius * 2 # 지름
radius3 = 2 * pi * radius # 원의 둘레
radius4 = pi * radius ** 2
radius5 = pi * radius * radius

print(radius2, radius3, radius4, radius5)

print("지름 :", radius2)
print("원의 둘레 :", radius3)
print("원의 넓이 :", radius4)

print(f"원의 넓이 : {radius4}")

# 변수를 선언해서 다음의 값을 출력해주세요
# 원의 지름
# 원의 둘레 : 2 파이 r
# 원의 넓이 : 파이 r^2
# -> 각자 변수명 선언 => 연산식을 활용해서 위의 값을 도출!! -> 출력
```
